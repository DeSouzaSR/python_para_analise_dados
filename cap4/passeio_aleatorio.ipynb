# Random Walk

Implementação do passeio aleatório, usanddo random do numpy.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Implementação em Python puro

```python
import random
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if  random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

```python
plt.plot(walk)
```

## Implementação com numpy

```python
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
```

## Simulando vários passeios aleatórios

```python
nwalks = 5000 # número de passeios aleatórios
steps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps))
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(1)
```

```python
x = np.arange(len(walks[:,0]))
plt.plot(x, walks[:,2])
plt.show()
```
