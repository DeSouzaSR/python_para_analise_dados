# Progamação orientada a array

Suponhamos que queiramos avaliar a função $\sqrt{x^2 + y^2}$ para uma grade regular de valores $x$ e $y$.

```python
import numpy as np
```

```python
points = np.arange(-5, 5, 0.01) # 1000 pontos, igualmente espaçados
```

```python
xs, ys = np.meshgrid(points, points)
```

```python
xs
```

```python
z = np.sqrt(xs**2 + ys**2)
```

```python
z
```

Gráfico dos resultados

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(z, cmap = plt.cm.gray)
plt.colorbar()
plt.show()
```

Avaliando a função $x^2 - y^2$

```python
w = xs**2 - ys**2
w
```

```python
plt.imshow(w, cmap = plt.cm.gray)
plt.colorbar()
plt.show()
```
